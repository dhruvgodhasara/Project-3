## Box plot

```python
# Box plot - it clearly shows that attchd garrage is higher values and than decreasing order 
UV=data['GarageType'].value_counts()
sns.barplot(x=UV.index , y=UV.values)
plt.title("Garage Distribution")
plt.xlabel("GarageType")
plt.ylabel("Count")
plt.show()

# Insights
#> it clearly shows that attached garage type is more count than other one.
```
