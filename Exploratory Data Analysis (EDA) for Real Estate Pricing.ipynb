## Outliers

```python
## Outliers Detection using IQR(Only for Numeric Columns)
numeric_cols=data.select_dtypes(include=['number']).columns
Q1=data[numeric_cols].quantile(0.25)
Q3=data[numeric_cols].quantile(0.75)
IQR= Q3-Q1

outliers=((data[numeric_cols]<(Q1-1.5*IQR))|(data[numeric_cols]>(Q3+1.5*IQR))).sum()
print("Outliers Count per Column:\n",outliers)
```

```python
# Outliers Detection using IQR (only for numeric columns)
# numeric_cols = data.select_dtypes(include=['number']).columns

Q1=data['LotFrontage'].quantile(0.25)
Q3=data['LotFrontage'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR

data_no_outliers=data[(data['LotFrontage']>= lower_bound) & (data['LotFrontage']<=upper_bound)]
print(data_no_outliers)
#outliers =((data[numeric_cols]<(Q1-1.5*IQR))|(data[numeric_cols]>(Q3+1.5*IQR))).sum()
#print("Outliers Count Per Column:\n",outliers)
```

# Data Visulization

### Heat Map

```python
# A heatmap is a great way to visualize correlation or intensity between variables in a tabular format,
# where values are represented by colors.
# The most common use in data science is for visualizing correlation matrices.

plt.figure(figsize=(25,25))
sns.heatmap(corrDf,annot=True,cmap='rainbow')

# Insights
#> 1 or -1 : Perfact correlation (Positive or Negative)
# 0 : No Correlation
# Positive values= Indicates that as one features increases, the other tends to increase as well.
# Negative values= Indicates that as one features increases, the other tends to decrease.
```

## Box Plot

```python
# A box plot is a great way to visualize the distribution of a dataset and spot outliers, median, and spread.
sns.boxplot(x=data['Street'],y=data['SalePrice']) 
# Insights 
# In this Box Plot We learn when Street near the road side is more sale price than far the road side and clearly shown outliers.
```

```python
# LotArea in SalePrice
LotArea_counts=data['LotArea'].value_counts().head(10)
plt.figure(figsize=(15,5))
LotArea_counts.plot(kind='bar',color='orange')
plt.title('LotArea')
plt.xlabel('SalePrice')
plt.ylabel('LotArea')
plt.xticks(rotation=0 , ha='right')
plt.show()

# Insights
# we have learn from this plot that higher the lotarea higher the saleprice
```

## Scatter Plot

```python
# A scatter plot is a fundamental visualization used to display the relationship between two numerical variables.
x= data['LotArea']
y=data['SalePrice']

plt.scatter(x,y)
plt.title('Scatter plot of LotArea Vs SalePrice')
plt.xlabel('SalePrice')
plt.ylabel('LotArea')
plt.show()

# Insights
# > we have learn from this plot that LotArea is directly proportional to the SalePrice.
```

## Histogram

```python
# A histogram is a graphical representation of the distribution of numerical data. 
variable=data['SalePrice']

sns.histplot(variable, bins=50 ,kde=True)
plt.title('Histogram of SalePrice')
plt.xlabel('SalePrice')
plt.ylabel('frequency')
plt.show()

# Insights
# > We have Learn from this plot that max sale price at 1.2lakhs around and spread over 7lakh and data positively 
# skewed because tail of the data is on right hand side.
```
