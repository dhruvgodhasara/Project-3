```python
# Load DataSet
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\DHRUV PATEL\Dhruv\dhruv\Dataset\housing_data.csv")
```

```python
categorical_columns = data.select_dtypes(include=['object', 'category']).columns.tolist()

print("Categorical columns:", categorical_columns)
```

```python
numerical_columns = data.select_dtypes(include=['number']).columns

print("Numerical columns:", numerical_columns)
```

```python
data.drop(['Unnamed: 0','Alley', 'MasVnrType', 'Electrical', 'GarageYrBlt'],axis='columns')
```

### Feature Engineering

```python
# add new collumns from existing collumns
```

```python
data['House']=data['SalePrice'] - data['LotArea']
```

```python
data.head()
```

```python
data.shape      #(rows , columns)
```

## Scatter Plot

```python
plt.scatter(x='SalePrice',y='House',data=data)
plt.title("Relationship between SalePrice and House")
plt.xlabel('Sale Price')
plt.ylabel('House')
plt.show()

# Insights 
#>> we have see that linear relationship with increase sale price with House
```

```python
data.columns
```

```python
### Calculate total area & price per sqr ft
data['TotalUsableArea']=data['GrLivArea']+data['TotalBsmtSF']+data['GarageArea']
data['PricePerTotalSqrFt']=data['SalePrice']/data['TotalUsableArea']
```

```python
data.sample(5)
```

### Bar Plot

```python
corr=data.corr(numeric_only=True)['SalePrice'].sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=corr.head(5),y=corr.head(5).index,palette='viridis')
plt.title('Top Features Correlated with SalePrice')
plt.xlabel('Correlation Coefficient')
plt.ylabel('Counts')
plt.show()

## Insights
#>> We clearly shows that more saleprice with respect to more Correlation Coefficient.
```

### Scatter Plot

```python
plt.figure(figsize=(8,5))
plt.scatter(data['OverallQual'],data['TotalUsableArea'],c=data['SalePrice'],cmap='viridis')
plt.colorbar(label='Sale Price')
plt.xlabel('Overall Quality')
plt.ylabel('Total Usable Area')
plt.title('Overall Quality vs Total Usable Area (Colored by Sale Price)')
plt.show()

# Insights scatter plot shows that Total Usable Area and Overall Quality have a positive linear relationship.
# we have see that area and quality increase ,the sale price also tends to increase.
# however , few outliers detected.
```

### Box Plot

```python
sns.boxplot(x=data['PricePerTotalSqrFt'])
plt.title('Boxplot of Price Per Total Square Foot')
plt.xlabel('Price Per Sqft')
plt.show()
```

```python
corre=data.corr(numeric_only=True)
top_corre=corre['SalePrice'].sort_values(ascending=False)
print(top_corre)
```

```python
data['HasFirePlace']=data['Fireplaces'].apply(lambda x:1 if x>0 else 0)

plt.figure(figsize=(7,5))
sns.boxplot(x='HasFirePlace',y='SalePrice',data=data)
plt.title('Impact of Fireplace on SalePrice')
plt.xticks([0,1],['No Fireplace','Has Fireplace'])
plt.xlabel('Fireplace')
plt.ylabel('SalePrice')
plt.tight_layout()
plt.show()

# Insight
# We have clearly shows that has fireplace is more saleprice than no fireplace.
```

```python
data['HasPool']=data['PoolArea'].apply(lambda x:1 if x>0 else 0)
```

```python
data['HasGarage']=data['GarageArea'].apply(lambda x: 1 if x > 0 else 0)
```

```python
amenties=['HasPool','HasGarage','HasFirePlace']
for amenity in amenties:
    print(f"\n {amenity} Effect on SalePrice:")
    print(data.groupby(amenity)['SalePrice'].describe()[['mean','count']])
```
